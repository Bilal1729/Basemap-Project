```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import numpy as np
```

```python
map1 = Basemap(projection='ortho', lat_0=50.0, lon_0=-100.0,resolution='l')
map1.drawcoastlines()
plt.show()
```

```python
map1.fillcontinents(color='coral')
map1.drawcoastlines()
map1.drawcountries()
plt.show()
```

```python
map = Basemap(projection='ortho',lat_0=50.0, lon_0=-100.0,resolution='l')
map.drawcoastlines()
map.drawcountries()
map.fillcontinents(color='coral')
map.drawmapboundary()
map.drawmeridians(np.arange(0, 360, 30))
map.drawparallels(np.arange(-90, 90, 30))
plt.show()
```

```python
map=Basemap(projection='merc',lat_0=57,lon_0=-135,
            resolution='h',area_thresh=0.1,
            llcrnrlon=76.25,llcrnrlat=8.083333333333334,
            urcrnrlon=80.33333333333333,urcrnrlat=13.583333333333334)
map.drawcoastlines()
map.drawcountries()
map.fillcontinents(color='green')
lons = 78.14189480516721
lats = 8.763713617974423
x,y = map(lons, lats)
map.plot(x, y, 'bo', markersize=10)
plt.text(x,y,'THOOTHUKUDI')
plt.show
```
